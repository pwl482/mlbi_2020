```python
from pyopenms import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
```

```python
Cl_L_L_dir = glob.glob("data/Class_L_L/*featureXML")
L_L_features = []
for file in Cl_L_L_dir:
    features = FeatureMap()
    FeatureXMLFile().load(file, features)
    rt_list = [] # retention time
    mz_list = [] # mass to charge
    for f in features:
        rt_list.append(f.getRT())
        mz_list.append(f.getMZ())
    heatmap, xedges, yedges = np.histogram2d(rt_list, mz_list, bins=25, range=[[0, 7000], [0, 1700]])
    L_L_features.append(heatmap.ravel())
len(L_L_features)
```

```python
Cl_L_0_dir = glob.glob("data/Class_L_0/*featureXML")
L_0_features = []
for file in Cl_L_0_dir:
    features = FeatureMap()
    FeatureXMLFile().load(file, features)
    features = FeatureMap()
    FeatureXMLFile().load(file, features)
    rt_list = [] # retention time
    mz_list = [] # mass to charge
    for f in features:
        rt_list.append(f.getRT())
        mz_list.append(f.getMZ())
    heatmap, xedges, yedges = np.histogram2d(rt_list, mz_list, bins=25,range=[[0, 7000], [0, 1700]])
    L_0_features.append(heatmap.ravel())
len(L_0_features)
```

```python
df_L_L = pd.DataFrame(L_L_features)
df_L_L.to_csv('L_L_features.csv', index=False)
df_L_0 = pd.DataFrame(L_0_features)
df_L_0.to_csv('L_0_features.csv', index=False)
```
